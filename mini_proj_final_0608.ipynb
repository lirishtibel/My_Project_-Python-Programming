```python
#Part II – The Graph class
#Task 1 – Define the class

class Graph():
    def __init__(self,name,nodes=[]):
        self.name= name
        self.nodes = {}        
        for i in range(len(nodes)):
            self.nodes[nodes[i].name]=nodes[i]
                        
    def __str__(self):
        print('Graph Name:',self.name)
        for n in self.nodes.values():
            n.__str__()
            
    def __len__(self):
        return len(self.nodes.keys())#returns the number of nodes in the graph
     
    def __contains__(self, key):
        if type(key)==str:
            if key in self.nodes.keys():
                return True
            else:
                return False
        elif isinstance(key, Node)==True:#equivalent to type(key)==Node
            if key in self.nodes.values():
                return True
            else:
                return False
        else:
            raise TypeError("Key type is: str or Node, "+ str(key) + " is not one of them!")
            
    def __getitem__(self, name):
        if self.__contains__(name):
            return self.nodes[name]            
        else:
            raise KeyError(""+ str(name) + " is not in the graph!")
        
    def __add__(self, other):
        
        new_graph = Graph(self.name + ' + ' + other.name, [])  #Creating new graph
        
        for k,v in self.nodes.items():      #appending first graph nodes into new graph
            new_graph.nodes[str(k)]=v
            
        for kk,vv in other.nodes.items():   # appending second graph nodes only if not exists in first graph
            if kk not in new_graph:
                new_graph.nodes[str(kk)]=vv
                
            else:
                for k in new_graph.nodes:
                    if kk == k:
                        node_to_update = Node(k)
                        node_to_update_from = Node(kk)
                        for ne_k,ne_v in other.nodes[kk].neighbors.items():
                            new_graph.nodes[k].n_update(ne_k,ne_v)
                
        return new_graph
       
    def update(self, node):
        if isinstance(node, Node):
            if self.__contains__(node):#node is in the graph
                for key in node.neighbors.keys():
                    if self.__contains__(key):#neighbor is in graph
                        node.neighbors[key]=max(self.get_edge_weight(node.name,key),node.__getitem__(key))
            else:
                self.nodes[node.name]=node #add new node to the graph
        else:
            return None          
     
    def remove_node(self, name):
        if self.__contains__(name):#check with name if node is in self
                 self.nodes.pop(name)
        else:
            return none
            #raise KeyError(""+ str(name) + " is not in this graph!")
            
    #This method should not remove edges, in which name is a neighbor of other nodes in the graph. 
    
    def is_edge(self, frm_name, to_name):
        if self.__contains__(frm_name) and self.__contains__(to_name):
            if to_name in self.nodes[frm_name].neighbors.keys():
                return True  
            else:
                return False
        else:
            return None
        
    def add_edge(self,frm_name,to_name,weight):
        if self.__contains__(frm_name) and self.__contains__(to_name):
            self.nodes[frm_name].update(to_name, weight)
        else: None
    
    def remove_edge(self, frm_name, to_name):
        if self.__contains__(frm_name):#the node is in graph
            if self.nodes[frm_name].__contains__(to_name)==True:#to_name is a neighbor of frm_name.
                self.nodes[frm_name].remove_neighbor(to_name)
            else:
                return print("{} is not a neighbor of the current Node {} !\n".format(to_name,frm_name))
        else:
            return print("{} is not in graph {} !\n".format(frm_name,self.name))   
        

    def get_edge_weight(self, frm_name, to_name):
        x=self.is_edge(frm_name, to_name)
        if x:
            return self.nodes[frm_name].neighbors.get(to_name)
        else:
            return x                                                        

    def get_path_weight(self,path=[]):
        if len(path)<1:
            return None
        else:
            Isfeasible=[]
            len_path=len(path)
            for n in range(0,len_path-1):
                Isfeasible.append(self.is_edge(path[n],path[n+1]))
            if all(Isfeasible):
                path_weight=0
                for i in range(0,len_path-1):
                    path_weight=path_weight+ self.get_edge_weight(path[i],path[i+1])
                return path_weight
            else:
                return None
            

    def is_reachable(self, frm_name, to_name):
        
        all_nodes =self.nodes[frm_name].neighbors.keys()
        previous = set()
        
        if frm_name in self and to_name in self:
            while  (len(all_nodes) != len(previous)) and to_name not in all_nodes:
                previous = all_nodes
                for n in all_nodes:
                    round_nodes =self.nodes[n].neighbors.keys()
                    all_nodes = all_nodes | round_nodes
            if to_name in all_nodes:
                return True
            else:
                return False            
         
    def find_all_paths(self, frm_name, to_name, path=[]):
            path = path + [frm_name]
            if frm_name == to_name:
                return [path]
            if frm_name not in self.nodes:
                return []

            paths = []

            for node in self.nodes[frm_name].neighbors:
                if node not in path:
                    newpaths = self.find_all_paths( node, to_name,path)
                    for newpath in newpaths:
                        paths.append(newpath)
                        return paths
                                            
    #find_shortest_path(self, frm_name, to_name) – returns the path from frm_name to
    #to_name which has the minimum total weight.
    def dijkstras(self,frm_name, to_name):
        x=[frm_name,to_name] 
        #if any ([vertax not in self.nodes.keys() for vertax in x]):
        if not self.is_reachable(frm_name, to_name):
            return None
          
        path = {}
        visited_list = []#{frm_name: 0}
        unvisited_list = set(self.nodes)
        # בונים דיקשנארי של כל הצמתים בגרף ומכניסים מרחק גבוה =  99999999 למעט לצומת התחלה שלה מכניסים מרחק 0 מעצמה
        for node in self.nodes:
            path[node]=999999999
        path[frm_name]=0
            
        min_node = frm_name  
        while len(unvisited_list)>0:
            unvisited_list.remove(min_node)
          
            for neighbor in  self.nodes[min_node].neighbors:
                if neighbor in visited_list or self.__contains__(neighbor)==False:
                    continue
                path_Temp=path[min_node] + self.get_edge_weight(min_node,neighbor)               
                if path_Temp < path[neighbor]:
                    path[neighbor]= path_Temp                    
        
            sorted_path = sorted(path.items(), key=lambda t: t[1])
        
            Temp_List = []
            for node in sorted_path:
                if node[0] not in unvisited_list:
                    continue
                Temp_List.append(node[0])
                break
            
            if len(Temp_List) > 0:
                min_node =  Temp_List[0]  
            if min_node not in visited_list:
                visited_list.append(min_node)
                
        return path[to_name]  
```

```python
#Part III – Non-directional graph

class NonDirectionalGraph(Graph):
    def __init__(self,name,nodes=[]):
        Graph.__init__(self,name,nodes)
        self.NonDirectional=True
        
    def __str__(self):
        print('NonDirectionalGraph Name:',self.name)
        for n in self.nodes.values():
            n.__str__()
      
    def add_edge(self,frm_name,to_name,weight=1):
        if self.__contains__(frm_name) and self.__contains__(to_name):
            self.nodes[frm_name].update(to_name, weight)
            self.nodes[to_name].update(frm_name, weight)
        else: None
    
    def remove_edge(self, frm_name, to_name):
        if self.__contains__(frm_name):#the node is in graph
            if self.nodes[frm_name].__contains__(to_name)==True:#to_name is a neighbor of frm_name.
                self.nodes[frm_name].remove_neighbor(to_name) 
                self.nodes[to_name].remove_neighbor(frm_name)
            else:
                return print("{} is not a neighbor of the current Node {} !\n".format(to_name,frm_name))
        else:
            return print("{} is not in graph {} !\n".format(frm_name,self.name))   
  
    
```
